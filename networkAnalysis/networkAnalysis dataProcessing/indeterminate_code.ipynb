```python
def calculate_group(row):
    if row['pvalues'] > 0.05:
        return 'Insignificant'
    elif row['pvalues'] < 0.05 and row['Odds Ratio'] > 1 and row['adjusted pval'] < 0.05:
        return 'High Risk'
    elif row['pvalues'] < 0.05 and row['Odds Ratio'] > 1 and row['adjusted pval'] >= 0.05:
        return 'Low Risk'
    elif row['pvalues'] < 0.05 and row['Odds Ratio'] < 1 and row['adjusted pval'] < 0.05:
        return 'High Protective'
    elif row['pvalues'] < 0.05 and row['Odds Ratio'] < 1 and row['adjusted pval'] >= 0.05:
        return 'Low Protective'
    return '-'
female_nodes['Group'] = female_nodes.apply(calculate_group, axis=1)
male_nodes['Group'] = male_nodes.apply(calculate_group, axis=1)
overall_nodes['Group'] = overall_nodes.apply(calculate_group, axis=1)
```
